```python
#importando a biblioteca pandas
import pandas as pd
```

```python
#trazendo o arquivo o phyton
aluguel_casa = pd.read_excel("houses_rent.xlsx")
```

```python
#exibindo as primeiras 10 linhas do arquivo
aluguel_casa.head(10)
```

```python
#exibindo as ultimas 5 linhas da tabela
aluguel_casa.tail()
```

```python
#alterando nome de algumas colunas
aluguel_casa = aluguel_casa.rename(columns={"parking spaces":"parking_spaces", "rent amount":"rent_amount", "property tax":"property_tax", "fire insurance":"fire_insurance"})
```

```python
#exibindo tabela com colunas alteradas
aluguel_casa.head()
```

```python
#verificando os tipos de dados
aluguel_casa.dtypes
```

```python
#mudando o o valor de aluguel de inteiro para float
aluguel_casa = aluguel_casa.astype({
    "rent_amount":"float",
    "property_tax":"float",
    "fire_insurance":"float",
    "total":"float"
    })
```

```python
#vizualizando como ficou
aluguel_casa.head()
```

```python
#formatando os dados float para melhorar a vizualizao de valores monetarios com 2 casas decimais
pd.options.display.float_format = '{:,.2f}'.format
aluguel_casa.head()
```

```python
#entendendo a funo .info - importante para ver a acuracidade dos dados
aluguel_casa.info()
```

```python
#mostrando apenas algumas colunas e bservando as suas estatisticas atraves do .describe (mean == media , std == desvio padro)
aluguel_casa[["area", "rooms", "bathroom", "parking_spaces", "rent_amount", "property_tax", "fire_insurance", "total"]].describe()
```

```python
aluguel_casa.head()
```

```python
# filtrando opes pois temos um budget de 4000 euros para gastar com alugel
aluguel_casa = aluguel_casa[(aluguel_casa["total"] <= 4000) & (aluguel_casa["animal"]=="accept") & ((aluguel_casa["rooms"]==2) | (aluguel_casa["rooms"]==3))]
```

```python
#filtrando de maneira mais simple com .query
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept' and (rooms == 2 or rooms == 3)")
aluguel_casa.head()
```

```python
#criando uma nova tabelinha agrupando a cidade e valor mdio e usando no final o .reset_index() para deixar em formato de data frame
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
#criando uma nova tabela agrupando qtde de banheiros e quartos e o valor mdio
agg_bath_room_price = aluguel_casa.groupby(["bathroom", "rooms"])["total"].agg("mean").reset_index()
agg_bath_room_price
```

**Entrando na parte visual usando a biblioteca Siborn**

```python
#criando um grafico de dispeo dos dados, comeando por criar uma nova tabela apenas para area
agg_area = aluguel_casa[["area"]]
```

```python
#filtrando os dados dentro da nova tabela de area
agg_area = agg_area.query("area <= 100")
```

```python
#verificando a qualidade dos daods filtrados
agg_area.describe()
```

```python
#crisndo um histograma para verificar como ficaram os dados em forma de grafico
sns.histplot(data=agg_area, x="area");
```

**Aps realizarmos todo este tratamento de dados e vizualizarmos alguns graficos para nos ajudar a vizualizar as melhores oportunidade de aluguel, vamos escolher o apartamento com base em nossa analise de dados.**

```python
#como as analises mostraram, na cidade PORTO os aluguel so mais baratos, portanto vamos filtrar a cidade PORTO
porto_data = aluguel_casa.query("city == 'Porto'")
porto_data.head()
```

```python
#analisando alguns parametros
porto_data.describe()
```

```python
#verificando quantos apartamentos existem como opoes por andar
porto_data.groupby("floor").index.nunique().sort_values(ascending=False)
```

```python
#agora vamos ver a mdia de preo por andar
#criando uma nova tabela agrupando
agg_floor_price = porto_data.groupby(["floor"])["total"].agg("mean").reset_index()
agg_floor_price
```

```python
#imprimindo a nova tabela em formato grafico de barras
sns.barplot(agg_floor_price, x="floor", y="total");
```

```python
#Analisando o grafico, e sabendo que andares mais altos valorizam mais, vamos filtrar criar uma nova tabela apenas com os apartamentos do decimo sexto andar
sixteenth_floor = porto_data.query("floor == 16")
sixteenth_floor
```
